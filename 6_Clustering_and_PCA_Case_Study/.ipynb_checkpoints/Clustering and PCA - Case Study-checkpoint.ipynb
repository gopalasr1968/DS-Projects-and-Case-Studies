> ## Problem Statement
>- An NGO Organization have raised fund of around `$10 millions` and they would like to spend the money strategically and effiectively , i.e., basically based on the countries that are in direst of needs.<br>
>- So, the aim is to categorize or cluster countries based on some socio-economic and health factors that determine the overall development of the country.<br>
>- The final goal is to suggest countries which they should focus most on the investment.<br>

> **Data Set provided has following details** ;
  1. country - Name of the country
  2. child_mort - Death of children under 5 years of age per 1000 live births
  3. exports - Export of goods and services. given as percentage of the total GDP
  4. health - Total health spending as percentage of Total GDP
  5. imports - Imports of goods and services. given as percentage of total GDP
  6. income - Net income per person
  7. inflation - The measurement of annual growth rate of the total GDP
  8. life_expec - The avg. number of year of new born child would live if the current mortality patterns are to remain same
  9. total_fer - The number of children that would born to each women if the current age fertility rates remains same
  10. GDPP - the GDP per capita.(calculated as the total GDP divided by total population)

### Step 7: Hopkins Statistics

>- The Hopkins statistic, is a statistic which gives a value which indicates the cluster tendency, in other words: how well the data can be clustered.

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan

def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hopkins(country_norm)
```

#### **Observations**<br>
>- As the Hopkins analysis above , the value are ~ above 0.70 indicates there is a high tendency of cluster.

### Step 8: K-Means Clustering

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 16):
    kmeans = KMeans(n_clusters=k).fit(df_pca_final5)
    sse_.append([k, silhouette_score(df_pca_final5, kmeans.labels_)])
```

#### Silhouette Analysis

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

#### Elbow Analysis -  Calculation of Sum of Squared Distances  

```python
# sum of squared distances
ssd = []
K= range(1,12)
for num_clusters in list(range(1,12)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(df_pca_final5)
    ssd.append(model_clus.inertia_)

plt.plot(K,ssd,'bx-')
plt.xlabel('k')
plt.ylabel('Sum of Squared Distances')
plt.title("Elbow Method for optimal K")
plt.show();
```

#### **Observations**<br>
>- From the silheoutte analysis and elbow analsyis the optimum cluster can be anywhere between 3 and 5.
> - Ideal optimum is cluster is 3,as from the elbow plot.
>- We will start k-meansof 6 intially and then with optimum 3.
